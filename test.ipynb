### well structured - an ode to databases


```python
import sqlite3
import pandas as pd
```

```python
df=pd.read_csv('test.csv')
```

```python
conn=sqlite3.connect('mytable.db')
```

```python
df.to_sql(name="employee",con=conn,if_exists='replace',index=False)
conn.commit()
```

```python
c = conn.cursor()
c.execute("SELECT * FROM employee")
rows = c.fetchall()
for row in rows:
    print(row)
```
